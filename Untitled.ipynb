```python
%matplotlib inline
import numpy as np # Import the numpy library, name it np
import cv2
from matplotlib import pyplot as plt

camera = cv2.imread("cameraman.tif") # Read the image from file
camera = cv2.cvtColor(camera, cv2.COLOR_BGR2RGB)
camera = cv2.cvtColor(camera, cv2.COLOR_RGB2GRAY)
plt.imshow(camera, cmap='gray') # Draw and show the image
plt.show()
```

```python

row,col= camera.shape
s_vs_p = 0.5
amount = 0.15
out = np.copy(camera)
# Salt mode
num_salt = np.ceil(amount * camera.size * s_vs_p)
coords = [np.random.randint(0, i - 1, int(num_salt))
      for i in camera.shape]
out[coords] = 255
# Pepper mode
num_pepper = np.ceil(amount* camera.size * (1. - s_vs_p))
coords = [np.random.randint(0, i - 1, int(num_pepper))
      for i in camera.shape]
out[coords] = 0
cv2.imwrite( "camera_s&p.png", out)
plt.imshow(out, cmap='gray') # Draw and show the filtered image
plt.show()
plt.imshow(camera, cmap='gray') # Draw and show the original image without noise
plt.show()
```

```python
def add_gaussian_noise(X_img):
    gaussian_noise_img = []
    row, col = X_img.shape
    # Gaussian distribution parameters
    mean = 0
    var = 0.1
    sigma = var ** 0.5
    
    gaussian = np.random.random((row, col, 1)).astype(np.float32)
    #gaussian = np.concatenate((gaussian, gaussian, gaussian), axis = 2)
    gaussian_img = cv2.addWeighted(X_img, 0.75, 0.25 * gaussian, 0.25, 0)

    gaussian_noise_img = np.array(gaussian_img, dtype = np.float32)
    return gaussian_noise_img
  
gaussian_noise_imgs = add_gaussian_noise(camera)
plt.imshow(gaussian_noise_imgs, cmap='gray') # Draw and show the filtered image
plt.show()
plt.imshow(camera, cmap='gray') # Draw and show the original image without noise
plt.show()
```

```python
row,col = camera.shape
mean = 0
sigma = 100
gauss = np.random.normal(mean,sigma,(row,col))
gauss = gauss.reshape(row,col)
noisy = camera + gauss
cv2.imwrite( "camera_gauss.png", noisy)
plt.imshow(noisy, cmap='gray') # Draw and show the filtered image
plt.show()
plt.imshow(camera, cmap='gray') # Draw and show the original image without noise
plt.show()
```
